```python
import numpy as np
import time

def calculator(n):
  Timex =[]
  for i in range(5):
   a = np.random.rand(n,n)
   b = np.random.rand(n,1)
   start = time.time()
   x = np.linalg.solve(a, b)
   end = time.time()
   Time = end - start
   Timex.append(Time)
   #print(f"matriz {i + 1} es: {np.allclose(np.dot(a, x), b)}")
   #if i == 4:
    #print(f"Fin del matrices de dimensiones {n}x{n}") #revisara que las soluciones sean correctas
  return Timex

def stats(n):
  result = calculator(n)
  mean = np.mean(result)
  std = np.std(result)
  #print(mean)
  #print(std)
  return mean
  return std

#si se quiere evitar el spam de verificar si las matrices son correctas
#comentar las lineas 14, 15 y 16, de lo contrario ejecutar tal y como esta
```

# Graficar


```python
import matplotlib.pyplot as plt
#graficar los promedios de tiempo normalizados(divididos) por el tiempo que tomo para la matriz 10x10
#incluir barras de error, las cuales serian 3 veces la desviacion estandar
n_values = [10, 100, 200, 500, 1000, 1200, 1500, 2000, 2500, 3000, 4000, 5000]

first_mean = stats(n_values[0])

# Calculate the mean execution time for each value of n
mean_execution_times = [stats(n)/ first_mean for n in n_values]
std_execution_times = [stats(n) for n in n_values]
std_execution_times = [3 * std for std in std_execution_times]

# Plot the mean execution time versus n
#plt.plot(n_values, mean_execution_times)
plt.errorbar(n_values, mean_execution_times, yerr=std_execution_times)
plt.xlabel('n')
plt.ylabel('Mean Execution Time')
plt.title('Mean Execution Time vs. n')
plt.show()
```
