```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
X_len=50
Y_len=50
a = np.linspace(0,4,100)
b = np.linspace(0,np.pi,100)
A, B = np.meshgrid(a, b)
Psi_sol = np.zeros_like(A)

Psi_sol[:, 0]=10
Psi_sol[:, -1]=150
Psi_sol[-1, 0:50]=150
Psi_sol[1:, 50:-1]= 0

deltar=1
deltatheta=1


for iterations in range(1000):
    for r in range(1, 4):
        for th in range(1, 4):
            Psi_sol[r][th]=((Psi_sol[r+1][th]+Psi_sol[r-1][th])/r+1)+((Psi_sol[r+1][th])/r*(r+1))+((Psi_sol[r][th+1]+Psi_sol[r][th-1])/r**2*(th+1)**2)

# Plot the heat map in polar coordinates

ax = plt.subplot(111, polar=True)
ax.set_yticklabels([])
ctf = ax.contourf(a, b, Psi_sol, cmap=cm.jet)
plt.colorbar(ctf)
plt.axis([0,2*np.pi,0,2.5])
plt.show()
```
